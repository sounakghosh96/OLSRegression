```python
import pandas as pd
df = pd.read_csv('Assignment2_data.csv')
```

```python
df = df[df['wave'] == 1981]
df = df.fillna(0)
```

```python
X = df[['sex','sch','annhrs','annlabinc','yrsexp','yrsExpSq']]
XVal = X.to_csv('XVal.csv')
X.loc[df["sex"] == 2, "sex"] = 0
X
```

```python
Y = df[['hrwage']]
Y
```

```python
import statsmodels.api as sm 
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, Y)
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
predictions = model.predict(X)
print_model = model.summary()
print(print_model)
```

```python
XMen = X[X['sex'] == 1]
XMen = XMen[['sch','annhrs','annlabinc','yrsexp','yrsExpSq']]
Men = df[df['sex'] == 1]
YMen = Men[['hrwage']]
YMen
```

```python
XFem = X[X['sex'] == 0]
XFem = XFem[['sch','annhrs','annlabinc','yrsexp','yrsExpSq']]
Fem = df[df['sex'] == 2]
YFem = Fem[['hrwage']]
YFem
```

```python
regrM = linear_model.LinearRegression()
regrM.fit(XMen, YMen)
XMen = sm.add_constant(XMen)
modelM = sm.OLS(YMen, XMen).fit()
predictions = modelM.predict(XMen)
print_modelM = modelM.summary()
print(print_modelM)
```

```python
regrF = linear_model.LinearRegression()
regrF.fit(XFem, YFem)
XFem = sm.add_constant(XFem)
modelF = sm.OLS(YFem, XFem).fit()
predictions = modelF.predict(XFem)
print_modelF = modelF.summary()
print(print_modelF)
```
